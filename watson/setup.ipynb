```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.info()

print('\nUnique Values:')
for col in df.columns:
    print('col:',col, df[col].unique())
```

```python
# convert to code
mylist = list(df.select_dtypes(include=['object']).columns)
#mylist.remove('customerId')
mylist.remove('customerID')
mylist.remove('TotalCharges')
mylist.remove('Churn')
df['churn'] = (df['Churn'] == 'Yes').astype(int)
df['TotalCharges'] = df['TotalCharges'].apply(pd.to_numeric, errors='coerce')

print('removing rows with NaN on TotalCharges:', len(df) - len(df[np.isfinite(df['TotalCharges'])]))
df = df[np.isfinite(df['TotalCharges'])]
df = df.drop(["customerID", "Churn"], axis=1)

print('converting to code:\n', mylist)

df = pd.get_dummies(df, columns=mylist)
df = df.astype(np.float64)
df.head()
```
